## Checking RMSE value for different models

```python
#Linear Model
import statsmodels.formula.api as smf 

linear_model = smf.ols('Sales~t',data=Coca_Cola_new).fit()
pred_linear =  pd.Series(linear_model.predict(pd.DataFrame(test['t'])))
rmse_linear = np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_linear))**2))
rmse_linear
```

```python
##Multiplicative Seasonality

Mul_sea = smf.ols('log_sales~Q1+Q2+Q3+Q4',data = Coca_Cola_new).fit()
pred_Mult_sea = pd.Series(Mul_sea.predict(test))
rmse_Mult_sea = np.sqrt(np.mean((np.array(test['log_sales'])-np.array(np.exp(pred_Mult_sea)))**2))
rmse_Mult_sea
```

```python
#Quadratic 

Quad = smf.ols('Sales~t+t_square',data=train).fit()
pred_Quad = pd.Series(Quad.predict(test[["t","t_square"]]))
rmse_Quad = np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_Quad))**2))
rmse_Quad
```

```python
#Additive seasonality 

add_sea = smf.ols('log_sales~Q1+Q2+Q3+Q4',data=train).fit()
pred_add_sea = pd.Series(add_sea.predict(test[['Q1','Q2','Q3','Q4']]))
rmse_add_sea = np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

```python
#Additive Seasonality Quadratic 

add_sea_Quad = smf.ols('log_sales~t+t_square+Q1+Q2+Q3+Q4',data=train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(test[['t','t_square','Q1','Q2','Q3','Q4']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(test['Sales'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad
```

```python
#Multiplicative Additive Seasonality 

Mul_Add_sea = smf.ols('log_sales~t+Q1+Q2+Q3+Q4',data = train).fit()
pred_Mult_add_sea = pd.Series(Mul_Add_sea.predict(test))
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(test['Sales'])-np.array(np.exp(pred_Mult_add_sea)))**2))
rmse_Mult_add_sea 
```

```python
#Compare the results 

data = {"MODEL":pd.Series(["rmse_linear","rmse_Mult_sea","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_add_sea "]),"RMSE_Values":pd.Series([rmse_linear,rmse_Mult_sea,rmse_Quad,rmse_add_sea,rmse_add_sea_quad,rmse_Mult_add_sea])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_Values'])
```

## Data driven forcasting technique

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import SimpleExpSmoothing # SES
from statsmodels.tsa.holtwinters import Holt # Holts Exponential Smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("CocaCola_Sales_Rawdata.csv")
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(12,4))
df.Sales.plot(label="org")
for i in range(2,10,2):
    df["Sales"].rolling(i).mean().plot(label=str(i))
plt.legend(loc='best')
```

```python
decompose_ts_add = seasonal_decompose(df.Sales,period=12)
decompose_ts_add.plot()
plt.show()
```

```python
import statsmodels.graphics.tsaplots as tsa_plots
tsa_plots.plot_acf(df.Sales,lags=12)
tsa_plots.plot_pacf(df.Sales,lags=12)
plt.show()
```

```python
Train = df.head(100)
Test = df.tail(20)
```

```python
def MAPE(pred,org):
    temp = np.abs((pred-org)/org)*100
    return np.mean(temp)
```

### Simple Exponential Smoothing Method

```python
ses_model = SimpleExpSmoothing(Train["Sales"]).fit(smoothing_level=0.2)
pred_ses = ses_model.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_ses,Test.Sales) 
```

### Holtz Method

```python
# Holt method 
hw_model = Holt(Train["Sales"]).fit(smoothing_level=0.8, smoothing_slope=0.2)
pred_hw = hw_model.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_hw,Test.Sales)
```

### Holtz Winter Exponential Smoothing

```python
hwe_model_add_add = ExponentialSmoothing(Train["Sales"],seasonal="add",trend="add",seasonal_periods=12).fit() #add the trend to the model
pred_hwe_add_add = hwe_model_add_add.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_hwe_add_add,Test.Sales) 
```

### Holtz Winter exponential smoothing with multiplicative seasonality and additive trend

```python
hwe_model_mul_add = ExponentialSmoothing(Train["Sales"],seasonal="mul",trend="add",seasonal_periods=12).fit() 
pred_hwe_mul_add = hwe_model_mul_add.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_hwe_mul_add,Test.Sales)
```

## Holtz Winter exponential smoothing with multiplicative seasonality and additive trend is a best model as compared to other models

```python
plt.figure(figsize=(14,8))
plt.plot(train.index, train['Sales'], label='train', color="black")
plt.plot(test.index, test['Sales'], label='test', color="blue")
plt.plot(pred_ses.index,pred_ses, label='SimpleExponential', color="Green")
plt.plot(pred_hw.index,pred_hw, label='Holts', color="Red")
plt.plot(pred_hwe_add_add.index,pred_hwe_add_add, label='Holts_Winter', color="Violet")
plt.plot(pred_hwe_mul_add.index,pred_hwe_mul_add, label='Holts_Winter+Mul_Add', color="Pink")
plt.legend(loc='best')
plt.title("Forcasted Coca_Cola_Data")
plt.show()
```
